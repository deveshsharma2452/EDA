## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Copy dataset to avoid original changes
df_clean = df.copy()

# Drop duplicates
df_clean.drop_duplicates(inplace=True)

# Handle missing values
df_clean.dropna(subset=['CustomerID', 'Description'], inplace=True)

# Convert CustomerID to integer
df_clean['CustomerID'] = df_clean['CustomerID'].astype(int)

# Convert InvoiceDate to datetime
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])

# Create new feature: Revenue
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice']

# Extract time features
df_clean['Year'] = df_clean['InvoiceDate'].dt.year
df_clean['Month'] = df_clean['InvoiceDate'].dt.month
df_clean['Day'] = df_clean['InvoiceDate'].dt.day
df_clean['Hour'] = df_clean['InvoiceDate'].dt.hour

# Remove negative quantities (returns) for clean analysis
df_clean = df_clean[df_clean['Quantity'] > 0]

# Remove negative or zero prices
df_clean = df_clean[df_clean['UnitPrice'] > 0]

df_clean.head()
````

### What all manipulations have you done and insights you found?

1. Removed duplicate rows to ensure data quality.
2. Handled missing values in CustomerID and Description columns.
3. Converted InvoiceDate into datetime format for time-based analysis.
4. Created a new feature Revenue = Quantity × UnitPrice.
5. Extracted useful time features like Year, Month, Day, and Hour.
6. Removed negative values representing returns or invalid transactions.

Insights:
- Dataset contains return transactions (negative quantities).
- Revenue is not directly given, so feature engineering is required.
- Data is time-dependent, making it suitable for trend analysis.
- CustomerID missing values indicate guest purchases or incomplete records.
